```python
type_of_file = 'csv'

def file_name_provider(input_path):
    ind_s=input_path.rindex('/') +1
    ind_e=input_path.rindex('.')
    input_file_name = input_path[ind_s :ind_e]
    output_path= input_path[:ind_s] + input_file_name + type_of_file +'.'+ type_of_file
    return output_path
```

```python
import xml.etree.ElementTree as ET
import json
import re
import pandas as pd

def striphtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('', data)

Json_file = 'parsed_data/Posts.json'
tree = ET.parse('stackexchange/Posts.xml') #<xml.etree.ElementTree.ElementTree object at 0x7fa569a8cd68>
root = tree.getroot()

final_list = []
for i in root:
    data = i.attrib
    final_list.append(data)
data_list = []
for i in root:
    data = i.attrib['Body']
    data = striphtml(data)
    data_list.append(data)
with open(Json_file, 'w') as fp:
    json.dump(final_list, fp)
df = pd.read_json('parsed_data/Posts.json')
df['Body'] = data_list
df.to_csv('parsed_data/post2.csv')
```

```python
Json_file = 'parsed_data/Json_files/Users.json'
tree = ET.parse('stackexchange/Users.xml') #<xml.etree.ElementTree.ElementTree object at 0x7fa569a8cd68>
root = tree.getroot()

final_list = []
for i in root:
    data = i.attrib
    final_list.append(data)

data_list = []
for i in root:
    data = i.attrib
    if 'AboutMe' in data:
        data['AboutMe'] = striphtml(data['AboutMe'])
    data_list.append(data)
    
with open(Json_file, 'w') as fp:
    json.dump(final_list, fp)
df = pd.read_json('parsed_data/Json_files/Users.json')
df.to_csv('parsed_data/Users.csv')
```
